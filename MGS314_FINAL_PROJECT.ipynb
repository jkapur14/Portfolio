```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import pandas_datareader.data as web
```

```python
start = dt.datetime(2021,1,1)
end = dt.datetime(2025,12,31)
#pick a style
style.use('ggplot')
```

```python
df = web.DataReader('CIFR', 'stooq', start, end)
df
```

```python
df2 = df
df2 = df2.iloc[::-1]
df2
df=df2
df
```

```python
#Export to a csv
df.to_csv('cifr.csv')
#Reduce decimal places for easier viewing
pd.set_option("display.precision", 2)
#Check the data
df = pd.read_csv('cifr.csv', parse_dates=True, index_col=0)
df
```

```python
df = pd.read_csv('cifr.csv', parse_dates=True, index_col=0)
```

```python
df['Close'].plot()
plt.show()
```

```python
ax1= plt.subplot2grid((8,1), (0,0), rowspan = 5, colspan =1)
ax2= plt.subplot2grid((8,1), (6,0), rowspan = 1, colspan =1)
df['100ma'] = df['Close'].rolling(window=100, min_periods = 0 ).mean()
```

```python
ax1.plot(df.index, df['Close'])
ax1.plot(df.index, df['100ma'])
ax2.bar(df.index, df['Volume'])
```

```python
#The size for our chart:
plt.figure(figsize = (12,6))
#Plotting Close and SMA lines:
ax1 = plt.subplot2grid((8,1),(0,0),rowspan=5,colspan=1)
ax2 = plt.subplot2grid((8,1),(6,0),rowspan=3,colspan=1, sharex=ax1)
ax1.plot(df.index,df['Close'], 'r',label='Adjusted Close')
ax1.plot(df.index,df['100ma'], 'g', label= "100 Moving Average")
plt.xlabel('Date')
#ax1.set_xlabel('Date')
ax1.set_ylabel('Closing price ($)')
ax1.set_title('Close and 100 day Moving Average for CIFR')
ax1.legend()
ax2.bar(df.index,df['Volume'], label = "Volume", color='b')
ax2.legend()
plt.show()
```
